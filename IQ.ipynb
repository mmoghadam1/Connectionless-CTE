```python
from collections import defaultdict
from collections import OrderedDict
from scipy.spatial import distance
import matplotlib.pyplot as plt
import numpy as np

c_aoa_data = np.loadtxt('data/conn_aoa.txt',delimiter=',')
c_aod_data = np.loadtxt('data/conn_aod.txt',delimiter=',')
cl_aoa_data = np.loadtxt('data/noconn_aoa.txt',delimiter=',')
cl_aod_data = np.loadtxt('data/noconn_aoa.txt',delimiter=',')
```

```python

c_close_data = np.loadtxt('data/newest.txt',delimiter=',')
c_close_dict = defaultdict(list)
for arr in c_close_data:
    ch = arr[0]
    c_close_dict[ch].append(arr[1:])
c_close_dict = OrderedDict(sorted(c_close_dict.items(), key=lambda t: t[0]))

c_close_mag = {}
c_close_phase = {}
c_close_phase2 = {}
for ch in c_close_dict:
    data = c_close_dict[ch]
    c_close_mag[ch] = [(sublist[::2]**2 + sublist[1::2]**2)**0.5 for sublist in data]
    c_close_phase[ch] = [np.arctan(sublist[1::2]/sublist[0::2]) for sublist in data]
    c_close_phase2[ch] = [np.arctan2(sublist[1::2],sublist[0::2]) for sublist in data]


c_far_data = np.loadtxt('data/far.txt',delimiter=',')
c_far_dict = defaultdict(list)
for arr in c_far_data:
    ch = arr[0]
    c_far_dict[ch].append(arr[1:])
c_far_dict = OrderedDict(sorted(c_far_dict.items(), key=lambda t: t[0]))

c_far_mag = {}
c_far_phase = {}
c_far_phase2 = {}
for ch in c_far_dict:
    data = c_far_dict[ch]
    c_far_mag[ch] = [(sublist[::2]**2 + sublist[1::2]**2)**0.5 for sublist in data]
    c_far_phase[ch] = [np.arctan(sublist[1::2]/sublist[0::2]) for sublist in data]
    c_far_phase2[ch] = [np.arctan2(sublist[1::2],sublist[0::2]) for sublist in data]
    
c_new_data = np.loadtxt('data/new.txt',delimiter=',')
c_new_dict = defaultdict(list)
for arr in c_new_data:
    ch = arr[0]
    c_new_dict[ch].append(arr[1:])
c_new_dict = OrderedDict(sorted(c_new_dict.items(), key=lambda t: t[0]))

c_new_mag = {}
c_new_phase = {}
c_new_phase2 = {}
for ch in c_new_dict:
    data = c_new_dict[ch]
    c_new_mag[ch] = [(sublist[::2]**2 + sublist[1::2]**2)**0.5 for sublist in data]
    c_new_phase[ch] = [np.arctan(sublist[1::2]/sublist[0::2]) for sublist in data]
    c_new_phase2[ch] = [np.arctan2(sublist[1::2],sublist[0::2]) for sublist in data]
```

```python
c_aoa_dict = defaultdict(list)
for arr in c_aoa_data:
    ch = arr[0]
    c_aoa_dict[ch].append(arr[1:])
c_aoa_dict = OrderedDict(sorted(c_aoa_dict.items(), key=lambda t: t[0]))
    
c_aod_dict = defaultdict(list)
for arr in c_aod_data:
    ch = arr[0]
    c_aod_dict[ch].append(arr[1:])
c_aod_dict =  OrderedDict(sorted(c_aod_dict.items(), key=lambda t: t[0]))  
    
cl_aoa_dict = defaultdict(list)
for arr in cl_aoa_data:
    ch = arr[0]
    cl_aoa_dict[ch].append(arr[1:])
cl_aoa_dict = OrderedDict(sorted(cl_aoa_dict.items(), key=lambda t: t[0]))
    
cl_aod_dict = defaultdict(list)
for arr in cl_aod_data:
    ch = arr[0]
    cl_aod_dict[ch].append(arr[1:])
cl_aod_dict = OrderedDict(sorted(cl_aod_dict.items(), key=lambda t: t[0]))
```

```python
c_aoa_mag = {}
c_aoa_phase = {}
c_aoa_phase2 = {}
for ch in c_aoa_dict:
    data = c_aoa_dict[ch]
    c_aoa_mag[ch] = [(sublist[::2]**2 + sublist[1::2]**2)**0.5 for sublist in data]
    c_aoa_phase[ch] = [np.arctan(sublist[1::2]/sublist[0::2]) for sublist in data]
    c_aoa_phase2[ch] = [np.arctan2(sublist[1::2],sublist[0::2]) for sublist in data]

c_aod_mag = {}
c_aod_phase = {}
c_aod_phase2 = {}
for ch in c_aod_dict:
    data = c_aod_dict[ch]
    c_aod_mag[ch] = [(sublist[::2]**2 + sublist[1::2]**2)**0.5 for sublist in data]
    c_aod_phase[ch] = [np.arctan(sublist[1::2]/sublist[0::2]) for sublist in data]
    c_aod_phase2[ch] = [np.arctan2(sublist[1::2],sublist[0::2]) for sublist in data]

cl_aoa_mag = {}
cl_aoa_phase = {}
cl_aoa_phase2 = {}
for ch in cl_aoa_dict:
    data = cl_aoa_dict[ch]
    cl_aoa_mag[ch] = [(sublist[::2]**2 + sublist[1::2]**2)**0.5 for sublist in data]
    cl_aoa_phase[ch] = [np.arctan(sublist[1::2]/sublist[0::2]) for sublist in data]
    cl_aoa_phase2[ch] = [np.arctan2(sublist[1::2],sublist[0::2]) for sublist in data]

cl_aod_mag = {}
cl_aod_phase = {}
cl_aod_phase2 = {}
for ch in cl_aod_dict:
    data = cl_aod_dict[ch]
    cl_aod_mag[ch] = [(sublist[::2]**2 + sublist[1::2]**2)**0.5 for sublist in data]
    cl_aod_phase[ch] = [np.arctan(sublist[1::2]/sublist[0::2]) for sublist in data]
    cl_aod_phase2[ch] = [np.arctan2(sublist[1::2],sublist[0::2]) for sublist in data]
    
```
